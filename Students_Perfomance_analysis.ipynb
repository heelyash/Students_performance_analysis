```python
#Before moving furthure as we are doing are doing supervised ML model so we have to decided our target variable from the above dataset so we
#considered the total_marks as the target variable we will predict the total_marks of the student on the basis of the other features.
```

```python
x = data.drop(columns=['total_marks'],axis=1)
```

```python
#As the x data have the multiple features so check whether they all are have relevance or importance in the data or not we can use the PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca
```

```python
pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
pca_x = pd.DataFrame(pca.fit_transform(x),columns=['PC1','PC2'])
```

```python
#Now we finally move towards the stage of building machine learing model so now we can seprete the data into the train and test part
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(pca_x,y,test_size=0.3,random_state = 2)
```

```python
x_train.shape,x_test.shape,y_test.shape,y_train.shape #This is the shape our seprate data on this basis of the train and test
```
